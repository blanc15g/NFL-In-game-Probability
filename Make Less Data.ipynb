```python
winners_dict = []
for year in range(2002, 2015):
    with open('nfl_results-master/nfl %s.csv' % year, 'r') as csv_file:
        reader = csv.DictReader(csv_file)
        for line in enumerate(reader):
            winners_dict.append(line[1])
```

```python
winners_dict[0]
```

```python
winners = {}
for game in winners_dict:
    kickoff = game['kickoff']
    date = ('%s/%s/' % (int(kickoff[5:7]), int(kickoff[8:10])))+kickoff[2:4]
    gameid1 = date + '_'+ game['visiting_team'] + game['home_team']
    winners[gameid1] = {game['home_team']: game['home_score'], game['visiting_team']: game['visitors_score']}
    #game['visiting_team'] if game['visitors_score'] > game['home_score'] else game['home_team']
    gameid2 = date + '_' + game['home_team'] +game['visiting_team']
    winners[gameid2] ={game['home_team']: game['home_score'], game['visiting_team']: game['visitors_score']}
 
    #game['visiting_team'] if game['visitors_score'] > game['home_score'] else game['home_team']
```

```python
allow = set((0,1,2,3,6,7,8))
new_games = []
for game in grouped_by_game.values():
    team_scores = collections.defaultdict(list)
    for play in game:
        team_scores[play['off']].append(int(play['offscore']))
        team_scores[play['def']].append(int(play['defscore']))
    all_passed = True
    for i in range(len(game) - 1):
        for val in team_scores.values():
            if (val[i+1] - val[i]) not in allow:
                all_passed = False
    if all_passed:
        new_games.append(game)
```
